(Speed calculation: min of fwd and bwd difference except at endpoints)

Potential outliers are identified by converting speed into a psuedo $Z$-score by subtracting the 1-day centered trimmed mean and dividing the result by the trimmed standard deviation. In both cases, the trimming functions removes the highest and lowest 10\% of the data before calculating the statistic. Observations with absolute $Z$-scores higher than 3 are flagged as potential outliers.

$$Z = \frac{|U| - \langle |U| \rangle_{0.1}}{\sigma_{0.1} |U|}$$

(Figure: speed, Z, qq-plot)

Next, we compare the difference in position and speed to a local natural cubic spline interpolation. The natural cubic spline interpolation is fit to data in a centered 2-day window surrounding a chosen observation. The interpolation is carried out simultaneously on $X$, $Y$, and $|U|$. The test statistics are the differences $\Delta \Delta X = \sqrt{(X - \hat X)^2 }, \Delta Y = \sqrt{(Y - \hat Y)^2}$ and $\Delta U = |U| - \hat{|U|}$.

In order to evaluate the test statistics, we take a random sample of $n=100$ dates and calculate $\Delta X$ and $\Delta U$ for each of these dates.
Examination of the resulting histograms shows that $\Delta U$ is approximately $t$-distributed. 
We calculate best-fit parameters for each buoy, then using the point-percentile function estimate lower and upper bounds such that the probability that $\Delta U$ is within the bounds is 1-$\alpha$ = 0.999. Observations outside of these empirical upper and lower bounds are flagged as outliers.

On the other hand, $\Delta X$ is strictly nonnegative and right-skewed, but the distribution is unclear.
We calculate the sample interquartile range (IQR) and median ($m$), and consider any value larger than $m + 2.5 \times IQR$ to be an outlier. 

If either $\Delta X$ or $\Delta U$ is flagged as an outlier, the observation is masked.

```python
%%time
results = []
for buoy in buoy_data:
    buoy_df = buoy_data[buoy]    
    n = min(len(buoy_df), 200)
    for date in buoy_df.sample(n=n, random_state=1251).index:
        margin = pd.to_timedelta('24H')
        if len(buoy_df.loc[slice(date-margin, date+margin)]) >= 24:
            test = clean.fit_splines(date, buoy_df.loc[slice(date-margin, date+margin)], 'x', 'y', 'speed')
            xerr = test.loc[date, 'x_hat'] - test.loc[date, 'x']
            yerr = test.loc[date, 'y_hat'] - test.loc[date, 'y']
            zerr = test.loc[date, 'speed_err']
            Z = buoy_df.loc[date, 'z']
            si = buoy_df.loc[date, 'sampling_interval']
            results.append([buoy, si, Z, np.round(xerr, 1), np.round(yerr, 1), np.round(zerr, 4)])
results_df = pd.DataFrame(results, columns=['BuoyID', 'SamplingInterval', 'Z', 'deltaX', 'deltaY', 'deltaU'])
```

```python
%%time
import scipy.stats as stats
alpha = 0.001

# This one is by buoyID. What about by time of year? By frequency?
t_thresholds = []
for name, group in results_df.groupby('SamplingInterval'):
    if len(group) > 100:
        thresh_list = []
        for var in ['deltaX', 'deltaY', 'deltaU']:
            df, loc, scale = stats.t.fit(stats.trimboth(group[var], 0.01))
            lower = stats.t.ppf(alpha, df, loc=loc, scale=scale)
            upper = stats.t.ppf(1-alpha, df, loc=loc, scale=scale) # X s.t. P(x < X) = q
            thresh_list += [df, loc, scale, lower, upper]
        t_thresholds.append([name] + thresh_list)
```

# Alternate method for choosing points to test
1. Calculate the speed anomaly compared to mean of a 12H centered moving window
2. Calculate the standard deviation of the speed anomaly using the 12H centered moving window
3. Mask data with absolute anomaly greater than 2 standard deviations
4. Re-calculate standard deviation with masked data
5. Flag data points above 3 standard deviations for review

The order that points are evaluated in matters, too. If we don't remove the largest anomalies first, then further points are going to be seen as anomalous since the fitted spline will try to pass through those points. I've set it up currently to only fit splines on points that have not been flagged.


```python
run_alternate = False
if run_alternate:
    for buoy in buoy_data:
        h12_anom = buoy_data[buoy].speed - buoy_data[buoy].speed.rolling('12H', center=True).mean()
        h12_stdev = h12_anom.rolling('12H', center=True).std()
        highvals = np.abs(h12_anom) > 2*h12_stdev
        h12_trimmed_stdev = h12_anom.where(~highvals).rolling('12H', center=True).std()
        buoy_data[buoy]['stdev_flag'] = np.abs(h12_anom) > 3*h12_trimmed_stdev

        buoy_data[buoy]['deltaX_flag'] = 0
        buoy_data[buoy]['deltaY_flag'] = 0    
        buoy_data[buoy]['deltaU_flag'] = 0
        buoy_data[buoy]['deltaX'] = np.nan
        buoy_data[buoy]['deltaY'] = np.nan
        buoy_data[buoy]['deltaU'] = np.nan

        idx = buoy_data[buoy].loc[buoy_data[buoy]['stdev_flag']].index

        for date in idx:

            check_data = buoy_data[buoy].loc[slice(date-margin, date+margin)].dropna(subset=['x', 'y', 'speed'])
            check_data = check_data.loc[(check_data.index == date) | ~check_data.stdev_flag]
            if len(check_data) > 24:
                check_df = clean.fit_splines(date, check_data, 'x', 'y', 'speed')
                check_df['y_err'] = check_df['y'] - check_df['y_hat']
                si = buoy_data[buoy].loc[date,'sampling_interval']
                if si not in [10, 30, 60, 120]:
                    si = adjust_si(si)
                if np.abs(check_df.loc[date, 'x_err']) > 6*t_thresholds.loc[si, 'scale_dX']:
                    buoy_data[buoy].loc[date, 'deltaX_flag'] = 1
                if np.abs(check_df.loc[date, 'y_err']) > 6*t_thresholds.loc[si, 'scale_dY']:
                    buoy_data[buoy].loc[date, 'deltaY_flag'] = 1
                if np.abs(check_df.loc[date, 'speed_err']) > 6*t_thresholds.loc[si, 'scale_dU']:
                    buoy_data[buoy].loc[date, 'deltaU_flag'] = 1


                buoy_data[buoy].loc[date, 'deltaU'] = check_df.loc[date, 'speed_err']
                buoy_data[buoy].loc[date, 'deltaX'] = check_df.loc[date, 'x_err']
                buoy_data[buoy].loc[date, 'deltaY'] = check_df.loc[date, 'y_err'] 
        if len(idx) > 0:
            print(buoy, len(idx), np.sum(buoy_data[buoy].deltaX_flag),
                  np.sum(buoy_data[buoy].deltaY_flag),
                  np.sum(buoy_data[buoy].deltaU_flag))
```

```python
fig, ax = pplt.subplots(width=8, height=3)
timeslice = slice('2019-12-01', '2020-01-05')
for buoy in buoy_data_interp:
    if buoy in co1_buoys:
        ax.plot(buoy_data[buoy].speed.loc[timeslice], marker='.', lw=1, markersize=2, color='k')        
        ax.plot(buoy_data_interp[buoy].speed.loc[timeslice], marker='.', lw=1, markersize=2)
    
ax.plot(co1_u.median(axis=1).loc[timeslice], lw=1)
```

```python
fig, ax = pplt.subplots(width=8, height=3)
timeslice = slice('2020-02-01', '2020-05-05')
for buoy in buoy_data_interp:
    if buoy in ['2019O1', '2019P112', '2019P113', '2019P114', '2019P119']:
        ax.plot(buoy_data[buoy].x.loc[timeslice], marker='.', lw=1, markersize=2, color='k')        
        ax.plot(buoy_data_interp[buoy].x.loc[timeslice], marker='.', lw=1, markersize=2)
```

```python
u_anom = (dn_u.loc[timeslice].T - co1_u.median(axis=1).loc[timeslice]).T
```

```python
timeslice=slice('2020-04-01', '2020-05-01')
fig, ax = pplt.subplots(width=8, height=3)
for col in u_anom.columns:
    ax.plot(u_anom[col], color='k', lw=1, alpha=0.2)
for buoy in ['2019P112', '2019P113', '2019P114', '2019P119']:
    buoy_df = buoy_data_interp[buoy]
    speed_anom = buoy_df['speed'] - co1_u.median(axis=1)
    ax.plot(speed_anom.loc[timeslice], label=buoy, marker='')
#ax.legend(ncols=1)
```
