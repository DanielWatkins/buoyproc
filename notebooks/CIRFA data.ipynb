```python
"""Reading and formatting the CIRFA2022 buoy data"""

import pandas as pd
import json
import numpy as np
import os
import sys
sys.path.append('../')
from src import cleaning
```

```python
files = os.listdir('../../data/CIRFA2022/')
```

```python
dataloc = '../../data/CIRFA2022/'
deployments = ['2022_04_24', '2022_04_25', '2022_04_30',
               '2022_05_06', '2022_04_26', '2022_05_02',
               '2022_05_05', '2022_05_04', '2022_05_03']
filepaths = []
for date in deployments:
    files = os.listdir(dataloc + date)
    for file in files:
        if 'CIRFA' in file:
            subfiles = os.listdir(dataloc + date + '/' + file)
            for sfile in subfiles:
                if 'trajectory' in sfile:
                    filepaths.append(os.path.join(dataloc, date, file, sfile))
```

```python
cirfa22 = {}
for path in filepaths:
    name = 'CIRFA22_' + path.split('_')[-2]
    df = pd.read_fwf(path, index_col=False, skiprows=1, header=None, names=['date', 'time', 'latitude', 'longitude'])
    df.index = [pd.to_datetime(d + ' ' + h) for d, h in zip(df['date'], df['time'])]
    cirfa22[name] = df.copy()
```

```python
import proplot as pplt
pplt.rc.reso = 'med'
import warnings
warnings.simplefilter('ignore')
fig, ax = pplt.subplots(proj='lcc', proj_kw={'lon_0': -10}, width=5)
ax.format(land=True, latlim=(70, 80), lonlim=(-20, 5))
for buoy in cirfa22:
    df = cirfa22[buoy]
    ax.plot(df.longitude, df.latitude, label=buoy)
ax.legend(loc='r', ncols=1)
```

```python
import proplot as pplt
pplt.rc.reso = 'med'
import warnings
warnings.simplefilter('ignore')

saveloc = '../../data/processed/cirfa2022/'
os.mkdir(saveloc)
fig, ax = pplt.subplots(width=6, height=6)
for buoy in cirfa22:
    df = cirfa22[buoy].copy()
    ax.plot(df.latitude, label=buoy, lw=0, marker='.', ms=5)
    df2 = cleaning.standard_qc(df.round(4), lat_range=(50, 90))
    if df2 is not None:
        ax.plot(df2.latitude.where(~df2.flag), lw=0, marker='.', ms=2, color='w', label='')
        cirfa22[buoy]['flag'] = df2['flag']
        cirfa22[buoy].where(~df2['flag']).dropna().loc[:, ['latitude', 'longitude']].to_csv(
            saveloc + buoy + '.csv')
ax.legend(loc='r', ncols=1)
```
